```python
import math
def euclid(p1,p2):
    if len(p1) != len(p2):
        print('Not possible')
        return
    dis = sum((x1-x2)**2 for x1,x2 in zip(p1,p2))
    return math.sqrt(dis)
print(euclid((0,0),(5,0)))
```

```python
def calbill(units):
    perunit = 0
    add = 0
    total = 0
    if units<100:
        perunit = 1.5
        add = 25
    elif 101 <= units <= 200:
        perunit = 2.5
        add = 50
    elif 201 <= units <= 300:
        perunit = 4
        add = 5
    elif 301 <= units <= 350:
        perunit = 7
        add = 100
    else:
        return 1500
    total = units*perunit
    total += add
    return total
units = int(input('Enter the units:'))
print('Total Bill for ',units,'is:', calbill(units))
    
```

```python
def sumprime(num):
    isp = []
    isp = [True]*(num+1)
    isp[0] = isp[1] = False
    sum = 0
    for i in range(2,num+1):
        if isp[i]:
            sum += i 
            for j in range(i*i,num+1,i):
                isp[j] = False
    print(sum)
sumprime(5)
```

```python
class BANK:
    def __init__(self,bname,numcust):
        self.bname = bname
        self.numcust = numcust
    def printdeets(self):
        print('Bank Name:',self.bname)
        print('Cust NUm:',self.numcust)

class GOVT_BANK(BANK):
    def __init__(self,bname,numcust,brname,ifsc):
        super().__init__(bname,numcust)
        self.brname = brname
        self.ifsc = ifsc
    def printdeets(self):
        super().printdeets()
        print('Branch Name:',self.brname)
        print('IFSC COde:',self.ifsc)

class PRIVATE_BANK(BANK):
    def __init__(self,bname,numcust,brname,ifsc):
        super().__init__(bname,numcust)
        self.brname = brname
        self.ifsc = ifsc
    def printdeets(self):
        super().printdeets()
        print('Branch Name:',self.brname)
        print('IFSC COde:',self.ifsc)

g1 = GOVT_BANK('SBI',123,'RAMAIAH','SBIRAM123')
g1.printdeets()
p1 = PRIVATE_BANK('CORP',456,'CHANDRA LAYOUT','CORPCHAND132')
p1.printdeets()
```

```python
class TIME:
    def __init__(self,hours,mins,secs):
        self.hours = hours
        self.mins = mins
        self.secs = secs

    def __str__(self):
        return f'{self.hours} hours,{self.mins} mins,{self.secs} secs'

    def __add__(self,other):
        total_secs = self.hours * 3600 + self.mins * 60 + self.secs
        total_secs += other.hours * 3600 + other.mins * 60 + other.secs
        hours2,rem = divmod(total_secs,3600)
        mins2,secs2 = divmod(rem,60)
        return TIME(hours2,mins2,secs2)

    def __sub__(self,other):
        total_secs = self.hours * 3600 + self.mins * 60 + self.secs
        total_secs -= other.hours * 3600 + other.mins * 60 + other.secs
        hours2,rem = divmod(abs(total_secs),3600)
        mins2,secs2 = divmod(rem,60)
        return TIME(hours2,mins2,secs2)

t1 = TIME(2,30,59)
t2 = TIME(2,30,11)
print(t1+t2)
print(t1-t2)
```

```python
studred = [
    {'name':'Mohit','Branch':'ISE','Sec':'B'},
    {'name':'DEF','Branch':'CSE','Sec':'B'},
    {'name':'GHI','Branch':'ISE','Sec':'C'},
]
def createfi(stud,path):
    with open(path,'w') as file:
        for i in stud:
            file.write(f"{i['name']} {i['Branch']} {i['Sec']}\n")
def search(path,branch):
    with open(path,'r') as file:
        print(f"{'Name'} {'Branch'} {'Sec'}")
        for line in file:
            name,b,s = line.split()
            if branch == b:
                print(f"{name} {b} {s}")

path = 'rec.txt'
createfi(studred,path)
search(path,'ISE')
```

```python
def sympal(string):
    l1 = len(string)//2
    l2 = len(string)
    for i in range(0,l1):
        if string[i].lower() != string[l2-i-1].lower() and string[i].lower() != string[i+l1].lower():
            return False
    return True
string = input('Enter a string')
print('Is the string a symmetric palindrome:',sympal(string))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def readfile(path):
    if path.endswith('.csv'):
        return pd.read_csv(path)
    if path.endswith('.xlsx'):
        return pd.read_excel(path)
    else:
        print('Invalid FIle')
        return

def calmean(data):
    return np.mean(data['Height'])
def calmed(data):
    return np.median(data['Height'])
def calstd(data):
    return np.std(data['Height'])
def calvar(data):
    return np.var(data['Height'])

def calslope(data,x_label,y_label):
    s,i,p,r,sa = stats.linregress(data[x_label],data[y_label])
    return s,i

def scatter(data,x_label,y_label):
    plt.scatter(data[x_label],data[y_label])
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(f"Scatter plot of {x_label} and {y_label}")
    plt.show()

def regress(data,x_label,y_label):
    s,i = calslope(data,x_label,y_label)
    plt.scatter(data[x_label],data[y_label])
    plt.plot(data[x_label],s*data[x_label]+i,label = 'Regression Line',color='black')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(f"Scatter plot of {x_label} and {y_label}")
    plt.legend()
    plt.show()

path = 'data1.csv'
x_label = 'Height'
y_label = 'Weight'
data = readfile(path)
scatter(data,x_label,y_label)
print('Mean:',calmean(data))
print('Median:',calmedian(data))
print('Variance:',calvar(data))
print('STD:',calstd(data))
print('Slope and intercept',calslope(data,x_label,y_label))
regress(data,x_label,y_label)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('result.csv')
print(data.head(10))
subjects = data.columns[1:-2]
passfailcount = {'Overall':{'Pass':0,'Fail':0}}

for subject in subjects:
    passfailcount[subject] = {'Pass':data[data[subject]>=40][subject].count(),
                              'Fail':data[data[subject]<40][subject].count()}
    passfailcount['Overall']['Pass'] += passfailcount[subject]['Pass']
    passfailcount['Overall']['Fail'] += passfailcount[subject]['Fail']

print('\nPass Fail Count')
print(pd.DataFrame(passfailcount))

sns.scatterplot(x='Science',y='Maths',hue='Results',data=data)
plt.title('Scatter plot of Science vs Maths')
plt.show()


passfailcountdf = pd.DataFrame(passfailcount).T
passfailcountdf.plot(kind='bar',stacked=True)
plt.title('Pass Fail Count')
plt.xlabel('Subjects')
plt.ylabel('Count')
plt.show()

ovaralldf = pd.DataFrame(passfailcount['Overall'],index=['Overall'])
ovaralldf.plot(kind='bar',stacked=True)
plt.title('Overall Pass Fail Count')
plt.xlabel('Total')
plt.ylabel('Count')
plt.show()

plt.hist(data['Total'],bins=20,color='skyblue',edgecolor='black')
plt.title('Histagram of total Marks')
plt.xlabel('Total Marks')
plt.ylabel('Frequency')
plt.show()
```
